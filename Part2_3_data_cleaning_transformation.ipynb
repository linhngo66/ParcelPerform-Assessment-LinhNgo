##### Handle duplicated rows

```python
#Drop duplicates
log_table = log_table.drop_duplicates().sort_values('parcel_id')
```

##### Find EDD in additional_params

```python
#Explore additional params to find EDD
list(log_table['additional_params'].unique())

#There are a few different key words
#new_parcel_expected_time_first_end
#old_parcel_expected_time_first_end
#new_parcel_expected_time_first_start
#old_parcel_expected_time_first_start
#new_parcel_expected_time_latest_start
#old_parcel_expected_time_latest_start

###Assuming EDD = new_parcel_expected_time_first_end
```

```python
#Check if there is any new_parcel_expected_time_latest_end
for i in list(log_table['additional_params'].unique()):
  if re.search('new_parcel_expected_time_latest_end', i) != None:
    print(i)
    break
  print('no key_word')


#no such key word
###Assuming EDD = new_parcel_expected_time_first_end
###For parcel_id with only EDD101, assuming EDD = log_timestamp
```

```python
log_table['new_parcel_expected_time_first_end'] = log_table['additional_params'].apply(lambda x: json.loads(x).get('new_parcel_expected_time_first_end'))
```

```python
log_table['edd'] = log_table['new_parcel_expected_time_first_end'].where(pd.notna(log_table['new_parcel_expected_time_first_end']), log_table['log_timestamp'])
```

```python
### For those parcel_id with EDD102, only take the lastest row based on log_timestamp, drop the rest
log_edd102 = log_table[log_table.duplicated(subset = 'parcel_id', keep = False)].sort_values('log_timestamp', ascending = False).drop_duplicates('parcel_id')

### For those parcel_id with EDD101 only, keep
log_edd101_index = log_table['parcel_id'].value_counts() == 1
log_edd101 = log_table[log_table['parcel_id'].isin(log_edd101_index[log_edd101_index].index)]
log_edd101

final_log_edd = pd.concat([log_edd101, log_edd102])
```

```python
final_log_edd[['new_parcel_expected_time_first_end', 'edd']] = final_log_edd[['new_parcel_expected_time_first_end', 'edd']].apply(pd.to_datetime, utc=True)
```

```python
final_log_edd
```

## Data Prep: Question 2 - Accuracy of EDD

**Final table should have the following columns:** \\
parcel_id | carrier_name | final_delivery_date | trade_line | edd | is_domestic_delivery| log_key | is_first_attempt | edd_accuracy (in days) \\

**Prep steps**
* Remove all parcels with is_delivered = False
* Remove all parcels with null final_delivery_date
* Left join parcel_table with log_table on parcel_id
* Validation: If EDD is before pickup date
* Create trade_line column and is_domestic delivery, is_first_attempt column

##### Remove all parcels with is_delivered = False & null final_delivery_date

```python
#Drop is_delivered = False & null final_delivery_date
parcel_table2 = parcel_table[(parcel_table['is_delivered'] == True) & (parcel_table['final_delivery_date'].notna())]
```

##### Left join parcel_table with log_table on parcel_id

```python
merge_parcel_log = parcel_table2.merge(final_log_edd[['parcel_id', 'log_key' ,'new_parcel_expected_time_first_end', 'edd']], on = 'parcel_id', how = 'left')
```

```python
#No null resulted from merge
merge_parcel_log.isnull().sum()
```

#####Validation: Is EDD before pickup date in any row?

```python
#Validation
merge_parcel_log['time_diff_pickup'] = (merge_parcel_log['edd'] - merge_parcel_log['picked_up_date']).dt.total_seconds()/(60*60*24)
merge_parcel_log[merge_parcel_log['time_diff_pickup'] < 0]
```

```python
#Drop all rows where pickup time is after EDD
merge_parcel_log = merge_parcel_log[merge_parcel_log['time_diff_pickup'] >= 0]
```

##### Calculate EDD accuracy

```python
#Calculate edd_accuracy
merge_parcel_log['edd_accuracy'] = (merge_parcel_log['edd']- merge_parcel_log['final_delivery_date']).dt.total_seconds()/(60*60*24)

plt.figure(figsize = (12, 6))
plt.hist(merge_parcel_log['edd_accuracy'], bins=20, color='skyblue', edgecolor='black')

# Customize the plot
plt.xlabel('EDD')
plt.ylabel('Frequency')
plt.title('EDD Accuracy')

# Show the plot
plt.show()
```

##### Add other new columns: Trade lane, Domestic delivery, First attempt

```python
#Trade_lane and Domestic delivery
merge_parcel_log['trade_lane'] = merge_parcel_log['origin_country'] + ' - ' + merge_parcel_log['destination_country']
merge_parcel_log['is_domestic_trade_lane'] = np.where(merge_parcel_log['origin_country'] == merge_parcel_log['destination_country'], True, False)
```

```python
#Is_first_attempt
merge_parcel_log['is_first_attempt'] = np.where(merge_parcel_log['first_attempt_date'] == merge_parcel_log['final_delivery_date'], True, False)
merge_parcel_log['is_first_attempt'].value_counts()
```

```python
final_edd_accuracy = merge_parcel_log[['parcel_id', 'carrier_name', 'origin_country', 'destination_country', 'trade_lane',
                                      'is_domestic_trade_lane', 'log_key', 'is_first_attempt', 'edd', 'edd_accuracy']]

final_edd_accuracy
```

##### Correlation

```python
import scipy.stats as stats
correlation_coefficient, p_value = stats.pointbiserialr(merge_parcel_log['is_first_attempt'], merge_parcel_log['edd_accuracy'])

# Output the point-biserial correlation coefficient
print("Point-biserial correlation coefficient:", correlation_coefficient)
print("P-value:", p_value)
```

##### Regression

```python
#One-hot encoding all categorical variables
dummy_var = pd.get_dummies(final_edd_accuracy, columns=['is_domestic_trade_lane', 'log_key', 'carrier_name', 'is_first_attempt'], drop_first=True)
dummy_var
```

```python
# Independent variables
X = dummy_var[['is_domestic_trade_lane_True', 'log_key_EDD102', 'carrier_name_Initech', 'carrier_name_Massive', 'is_first_attempt_True']]

#Dependent variables
y = dummy_var['edd_accuracy']

# Constant term for the intercept
X = sm.add_constant(X)
```

```python
# Create the multiple linear regression
model = sm.OLS(y, X).fit()

# Summary
print(model.summary())

#R-squared is low, not a lot of changes in Y is explained by X
#P-value < .05 in log_key, carrier_name, first_attempt
#Such that log_key: Change of log_key (EDD102) makes EDD less accurate (+4.25)
#Carrier name has a significant effect on EDD accuracy, meaning certain carrier (...) is better at estimating than the other
#First attempt has less accurate estimation(increase estimation, not in line with the observation)??? Kinda question this???
```

## Data Prep: Question 3 - Further Exploration: Operation
**5 stages of delivery**
* not picked up
* picked up
* out for delivery
* first attempt
* delivered

### Time between stages

```python
parcel_ops = parcel_table
```

```python
parcel_ops['time_to_out_for_delivery'] = (parcel_ops['out_for_delivery_date'] - parcel_ops['picked_up_date']).dt.total_seconds()/(60*60*24)
parcel_ops['time_to_first_attempt'] = (parcel_ops['first_attempt_date'] - parcel_ops['out_for_delivery_date']).dt.total_seconds()/(60*60*24)
parcel_ops['time_to_final_delivery'] = (parcel_ops['final_delivery_date'] - parcel_ops['first_attempt_date']).dt.total_seconds()/(60*60*24)
parcel_ops['transit_time'] = (parcel_ops['final_delivery_date'] - parcel_ops['picked_up_date']).dt.total_seconds()/(60*60*24)

parcel_ops
```

```python
time_between_stages = ['time_to_out_for_delivery', 'time_to_first_attempt', 'time_to_final_delivery', 'transit_time']

for i in time_between_stages:
  print(parcel_ops[i].describe())

  plt.figure(figsize = (12, 6))
  plt.hist(parcel_ops[i], bins=50, color='skyblue', edgecolor='black')

  # Customize the plot
  plt.xlabel('Days')
  plt.ylabel('Frequency')
  plt.title(i)

  # Show the plot
  plt.show()
```

```python
### Handle negative values
#For time_out_for_delivery, distribution is pretty normal so we're replacing negative values with mean
#For the rest, we're replacing negative values with median

for i in time_between_stages:
  if i == 'time_to_out_for_delivery':
    parcel_ops[parcel_ops[i] < 0][i] = parcel_ops[i].mean()
  else:
    parcel_ops[parcel_ops[i] < 0][i] = parcel_ops[i].median()

parcel_ops
```

### Stage of parcel

```python
def stage_of_parcel(row):
  if row['is_delivered'] == True:
    return 'delivered'

  if row['is_delivered'] == False:
    if pd.notnull(row['first_attempt_date']):
      return 'first attempt'
    elif pd.notnull(row['out_for_delivery_date']):
      return 'out for delivery'
    elif pd.notnull(row['picked_up_date']):
      return 'picked up'
    else:
      return 'not picked up'

parcel_ops['stage_of_parcel'] = parcel_ops.apply(stage_of_parcel, axis = 1)
parcel_ops
```

```python
parcel_ops['stage_of_parcel'].value_counts()
```

```python
#Is_first_attempt
parcel_ops['is_first_attempt'] = np.where(parcel_ops['first_attempt_date'] == parcel_ops['final_delivery_date'], True, False)
```

```python
#Trade_lane and Domestic delivery
parcel_ops['trade_lane'] = parcel_ops['origin_country'].astype(str) + ' - ' + parcel_ops['destination_country'].astype(str)
parcel_ops['is_domestic_trade_lane'] = np.where(parcel_ops['origin_country'] == parcel_ops['destination_country'], True, False)

parcel_ops
```

```python
parcel_ops_col = ['parcel_id', 'carrier_name', 'time_to_out_for_delivery',
                  'time_to_first_attempt', 'time_to_final_delivery', 'transit_time',
                  'stage_of_parcel', 'trade_lane', 'is_domestic_trade_lane', 'is_first_attempt']
final_parcel_ops = parcel_ops[parcel_ops_col]
```
